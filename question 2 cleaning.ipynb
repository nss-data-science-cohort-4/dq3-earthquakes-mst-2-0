```python
import requests
import pandas as pd
from io import StringIO
import json 
import csv 
```

```python
# capture data within the tn boarders with longitude and latitude and convert to csv and start the time from 1900 to now
url = 'https://earthquake.usgs.gov/fdsnws/event/1/query?format=csv&minlatitude=34&maxlatitude=37&minlongitude=-92&maxlongitude=-84&starttime=1900-01-01&endtime=2020-10-17'
```

```python
r = requests.get(url)
```

```python
print(r.text[:1000])
```

```python
tn_quake_df = pd.read_csv(StringIO(r.text))
tn_quake_df.head()
```

```python
# capture the last word in the string wich is the state
tn_quake_df['state'] = tn_quake_df['place'].str[-9:]
```

```python
# capture the date  
tn_quake_df['date'] = tn_quake_df['time'].str[:10]
```

```python
# capture the time  
tn_quake_df['time'] = tn_quake_df['time'].str[11:]
```

```python
# capture the time  
tn_quake_df['time'] = tn_quake_df['time'].str[:8]
```

```python
# capture the year  
tn_quake_df['year'] = tn_quake_df['date'].str[:4]
```

```python
tn_quake_df.tail(10)
```

```python
tn_quake_df['state'].unique()
```

```python
# remove data that is not Tennessee
tn_quake_df = tn_quake_df.groupby('state').get_group('Tennessee')
```

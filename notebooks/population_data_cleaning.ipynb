```python
import pandas as pd
import numpy as np
```

```python
population_df = pd.read_csv('../data/pop_data.csv', header=4)
```

```python
population_df.head()
```

```python
china_pop = population_df[population_df['Country Name'] == 'China']
```

```python
china_pop.head()
```

```python
china_pop.drop(['Country Code', 'Indicator Name', 'Indicator Code'], axis=1, inplace=True)
```

```python
china_pop
```

```python
china_pop = pd.melt(china_pop, id_vars=['Country Name'], var_name='year', value_name='pop')
```

```python
china_pop.info()
```

```python
china_pop = china_pop.dropna()
```

```python
china_pop.to_csv('../data/china_pop_clean.csv')
```
